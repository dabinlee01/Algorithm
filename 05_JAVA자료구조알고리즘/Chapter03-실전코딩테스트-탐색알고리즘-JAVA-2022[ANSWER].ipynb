````java
import java.io.BufferedReader;
import java.io.ByteArrayInputStream;
import java.io.IOException;
import java.io.InputStreamReader;
import java.util.StringTokenizer;

public class Main {
    public static void main(String[] args) throws IOException {
        String inputData = """
        3
        10 1
        20 3
        30 4
        """;
        System.setIn(new ByteArrayInputStream(inputData.getBytes()));

        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));
        int T = Integer.parseInt(br.readLine());

        for (int i = 1; i <= T; i++) {
            StringTokenizer st = new StringTokenizer(br.readLine());
            int N = Integer.parseInt(st.nextToken());
            int M = Integer.parseInt(st.nextToken());
            int count = 0;
            for (int a = 1; a < N; a++) {
                for (int b = (a + 1); b < N; b++) {
                    if (((a * a) + (b * b) + M) % (a * b) == 0) {
                        count++;
                    }
                }
            }
            System.out.println(count);
        }
    }
}
````

````java
import java.io.BufferedReader;
import java.io.ByteArrayInputStream;
import java.io.IOException;
import java.io.InputStreamReader;

public class Main {
    public static void main(String[] args) throws IOException {
        
        String inputData = """
        3
        10
        20
        1000
        """;
        System.setIn(new ByteArrayInputStream(inputData.getBytes()));

        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));

        int T = Integer.parseInt(br.readLine());
        int[] triNum = new int[45];
        for(int i = 1; i < 45; i++) {
            triNum[i] = i * (i + 1) / 2;
        }
        for(int i = 0; i < T; i++) {
            int result = eureka(Integer.parseInt(br.readLine()), triNum);
            System.out.println(result);
        }
    }
    public static int eureka(int n, int[] triNum) {
        for(int i = 1; i < 45; i++) {
            for (int j = 1; j < 45; j++) {
                for (int k = 1; k < 45; k++) {
                    if ((triNum[i] + triNum[j] + triNum[k]) == n) {
                        return 1;
                    }
                }
            }
        }
        return 0;
    }
}
````

````java
import java.io.BufferedReader;
import java.io.ByteArrayInputStream;
import java.io.IOException;
import java.io.InputStreamReader;
import java.util.Arrays;

public class Main {
    public static void main(String[] args) throws IOException {
        String inputData = """
        20
        7
        23
        19
        10
        15
        25
        8
        13
        """;
        System.setIn(new ByteArrayInputStream(inputData.getBytes()));

        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));
        int[] height = new int[9];
        int sum = 0;
        for (int i = 0; i < height.length; i++) {
            height[i] = Integer.parseInt(br.readLine());
            sum += height[i];
        }
        Arrays.sort(height);

        int fake1 = 0;
        int fake2 = 0;
        for (int i = 0; i < height.length - 1; i++) {
            for (int j = i + 1; j < height.length; j++) {
                if ((sum - height[i] - height[j]) == 100) {
                    fake1 = i;
                    fake2 = j;
                    break;
                }
            }
        }

        for (int i = 0; i < height.length; i++) {
            if ((i == fake1) || (i == fake2)) {
                continue;
            }
            System.out.println(height[i]);
        }
    }
}
````

### 5. 연습문제: 예산 https://www.acmicpc.net/problem/2512
<div class="alert alert-block" style="border: 2px solid #E65100;background-color:#FFF3E0;padding:10px">
<font size="4em" style="color:#BF360C;">이진 탐색 문제 예</font>
</div>
<div class="alert alert-block" style="border: 2px solid #E65100;background-color:#FFF3E0;padding:10px">
<font size="4em" style="color:#BF360C;">완전 탐색 (Brute-force) 로 접근하면 안되는 이유: 시간 초과 </font>
</div>

````java
import java.io.BufferedReader;
import java.io.ByteArrayInputStream;
import java.io.IOException;
import java.io.InputStreamReader;
import java.util.Arrays;
import java.util.StringTokenizer;

public class Main {
    public static void main(String[] args) throws IOException {
        String inputData = """
        4
        120 110 140 150
        485
        """;
        System.setIn(new ByteArrayInputStream(inputData.getBytes()));

        // 상한액을 1부터 10만까지 1씩 늘려나가면서 합을 1만번 구할 경우 O(10만*1만=10억)으로 시간초과
       // 따라서 상한액을 1부터 10만 사이에서 이분탐색으로 구하면 O(log(10만)*1만)
        
        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));
        int N = Integer.parseInt(br.readLine());
        StringTokenizer st = new StringTokenizer(br.readLine());
        int[] arr = new int[N];
        int sum = 0;
        int left = 0;
        int right = 0;

        for(int i = 0; i < N; i++) {
            arr[i] = Integer.parseInt(st.nextToken());
            sum += arr[i];
        }
        int M = Integer.parseInt(br.readLine());
        Arrays.sort(arr);
        right = arr[N - 1]; // max

        while(left <= right) {
            int mid = (left + right) / 2;
            long budget =0;
            for (int i = 0; i < N; i++) {
                if (arr[i] > mid) {
                    budget += mid;
                } else {
                    budget += arr[i];
                }
            }
            if (budget <= M) {
                left = mid + 1;
            } else {
                right = mid - 1;
            }
        }
        System.out.println(right);
    }
}
````

<font size="4em" style="color:#BF360C;">완전 탐색 (Brute-force) 로 접근하면 안되는 이유: 시간 초과 (O(n^2)) </font>
</div>

````java
import java.io.BufferedReader;
import java.io.ByteArrayInputStream;
import java.io.IOException;
import java.io.InputStreamReader;
import java.util.ArrayList;
import java.util.StringTokenizer;

public class Main {
    public static void main(String[] args) throws IOException {
        String inputData = """
        5
        4 1 5 2 3
        5
        1 3 7 9 5
        """;
        System.setIn(new ByteArrayInputStream(inputData.getBytes()));

        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));
        int N = Integer.parseInt(br.readLine());
        ArrayList<Integer> digits = new ArrayList<Integer>();
        ArrayList<Integer> finds = new ArrayList<Integer>();

        StringTokenizer st = new StringTokenizer(br.readLine());
        for(int i = 0; i < N; i++) {
            digits.add(Integer.parseInt(st.nextToken()));
        }

        int M = Integer.parseInt(br.readLine());
        st = new StringTokenizer(br.readLine());
        for(int i = 0; i < N; i++) {
            finds.add(Integer.parseInt(st.nextToken()));
        }

        for (int i = 0; i < finds.size(); i++) {
            boolean find = false;
            for (int j = 0; j < digits.size(); j++) {
                if (finds.get(i) == digits.get(j)) {
                    System.out.println(1);
                    find = true;
                    break;
                }
            }
            if (!find) {
                System.out.println(0);
            }
        }
    }
}
Main.main(new String[0]);
````

````java
import java.io.BufferedReader;
import java.io.ByteArrayInputStream;
import java.io.IOException;
import java.io.InputStreamReader;
import java.util.ArrayList;
import java.util.Collections;
import java.util.StringTokenizer;

public class Main {
    public static void main(String[] args) throws IOException {

        String inputData = """
        5
        4 1 5 2 3
        5
        1 3 7 9 5
        """;
        System.setIn(new ByteArrayInputStream(inputData.getBytes()));

        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));
        int N = Integer.parseInt(br.readLine());
        ArrayList<Integer> arr = new ArrayList<Integer>();

        StringTokenizer st = new StringTokenizer(br.readLine());
        for(int i = 0; i < N; i++) {
            arr.add(Integer.parseInt(st.nextToken()));
        }

        Collections.sort(arr);
        int M = Integer.parseInt(br.readLine());
        st = new StringTokenizer(br.readLine());

        for(int i = 0; i < M; i++) {
            if(Collections.binarySearch(arr, Integer.parseInt(st.nextToken())) >= 0) {
                System.out.println(1);
            }
            else {
                System.out.println(0);
            }
        }
    }
}
Main.main(new String[0]);
````

<font size="4em" style="color:#BF360C;">이진 탐색으로 풀지 않으면 시간 초과 </foynt>

````java
import java.io.BufferedReader;
import java.io.ByteArrayInputStream;
import java.io.IOException;
import java.io.InputStreamReader;
import java.util.Arrays;
import java.util.StringTokenizer;

public class Main {
    public static void main(String[] args) throws IOException {
        String inputData = """
        4 7
        20 15 10 17
        """;
        System.setIn(new ByteArrayInputStream(inputData.getBytes()));

        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));
        StringTokenizer st = new StringTokenizer(br.readLine());
        int N = Integer.parseInt(st.nextToken());
        int M = Integer.parseInt(st.nextToken());

        int[] trees = new int[N];
        int start = 0;
        int end = 0;

        st = new StringTokenizer(br.readLine());
        for (int i = 0; i < trees.length; i++) {
            trees[i] = Integer.parseInt(st.nextToken());
            end = Math.max(end, trees[i]);
        }
        Arrays.sort(trees);

        int result = 0;
        while (start <= end) {
            long total = 0;
            int mid = (start + end) / 2;
            for (int i = 0; i < N; i++) {
                if (trees[i] > mid) {
                    total += trees[i] - mid;
                }
            }
            if (total < M) {
                end = mid - 1;
            } else {
                start = mid + 1;
                result = mid;
            }
        }
        System.out.println(result);
    }
}
````

### 8. 공유기 설치 https://www.acmicpc.net/problem/2110

<div class="alert alert-block" style="border: 2px solid #E65100;background-color:#FFF3E0;padding:10px">
<font size="4em" style="color:#BF360C;">이진 탐색 문제 예</font>
</div>
<div class="alert alert-block" style="border: 2px solid #E65100;background-color:#FFF3E0;padding:10px">
<font size="4em" style="color:#BF360C;">이진 탐색으로 풀지 않으면 시간 초과 </foynt>
</div>

````java
import java.io.BufferedReader;
import java.io.ByteArrayInputStream;
import java.io.IOException;
import java.io.InputStreamReader;
import java.util.Arrays;
import java.util.StringTokenizer;

public class Main {
    public static void main(String[] args) throws IOException {
        String inputData = """
        5 3
        1
        2
        8
        4
        9
        """;
        System.setIn(new ByteArrayInputStream(inputData.getBytes()));

        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));
        StringTokenizer st = new StringTokenizer(br.readLine());
        int N = Integer.parseInt(st.nextToken());
        int C = Integer.parseInt(st.nextToken());

        int[] houses = new int[N];
        for (int i = 0; i < N; i++) {
            houses[i] = Integer.parseInt(br.readLine());
        }
        Arrays.sort(houses);

        int start = 1;
        int end = houses[N - 1] - houses[0];
        int result = 0;

        while (start <= end) {
            int mid = (start + end) / 2; // 정답 후보
            int count = 1;
            int last = houses[0];

            for (int i = 0; i < N; i++) {
                if (houses[i] - last >= mid) {
                    count++;
                    last = houses[i];
                }
            }
            if (C > count) {
                end = mid - 1;
            } else {
                start = mid + 1;
                result = mid;
            }
        }
        System.out.println(result);
    }
}
````

- 투 포인터 알고리즘
   - 리스트에 순차적으로 접근해야 할 때, 두 개의 점(보통 start, end 또는 left, right)의 위치를 기록하며 처리하는 알고리즘
   - 이를 통해 보통 $O(N^2)$ 이상으로 해결해야하는 문제를 $O(N)$ 으로 해결해주는 문제
   - 이미 익힌 병합정렬 알고리즘이 투포인터 알고리즘을 기반으로 만들어졌다고 보면 됨
- 투 포인터는 최근 코딩 테스트에서 많이 나왔던 유형 
   - 유명한 유형이 아님에도, 최근 대기업, IT 스타트업 코딩 테스트에서 나온 케이스가 많았음
- 투 포인터는 다양한 변형이 있을 수 있지만, 기본적으로는 연속되는 value들을 이용하여 특정 목표에 맞는 값을 찾아주는 개념
  - 투 포인터는 비교적 간단한 알고리즘이지만 강력한 개념
     - 특별한 알고리즘 유형이라고 보기보다는 개념이 더 알맞아보임
     - 두 개의 인덱스를 사용해서 답을 구하는 전략  

#### $O(N^2)$ 알고리즘 (시간 초과)

````java
import java.io.BufferedReader;
import java.io.ByteArrayInputStream;
import java.io.InputStreamReader;
import java.util.StringTokenizer;

public class Main {
    public static void main(String[] args) throws Exception {
        String inputData = """
        4 2
        1 1 1 1
        """;
        System.setIn(new ByteArrayInputStream(inputData.getBytes()));

        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));
        StringTokenizer st = null;

        st = new StringTokenizer(br.readLine().trim());

        int N = Integer.parseInt(st.nextToken());
        int M = Integer.parseInt(st.nextToken());

        st = new StringTokenizer(br.readLine().trim());
        int[] A = new int[N];
        for (int i = 0; i < N; i++) {
            A[i] = Integer.parseInt(st.nextToken());
        }

        int sum = 0;
        int count = 0;

        for (int i = 0; i < N; i++) {
            for (int j = i; j < N; j++) {
                sum += A[j];
                if (sum == M) {
                    count++;
                    break;
                }
            }
            sum = 0;
        }

        System.out.println(count);
    }
}
````

#### 투 포인터 알고리즘 $O(N)$ (성공)

````java
import java.io.BufferedReader;
import java.io.ByteArrayInputStream;
import java.io.InputStreamReader;
import java.util.StringTokenizer;

public class Main {
    public static void main(String[] args) throws Exception {
        String inputData = """
                        10 5
                               1 2 3 4 2 5 3 1 1 2
                """;
        System.setIn(new ByteArrayInputStream(inputData.getBytes()));

        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));
        StringTokenizer st = null;

        st = new StringTokenizer(br.readLine().trim());

        int N = Integer.parseInt(st.nextToken());
        int M = Integer.parseInt(st.nextToken());

        st = new StringTokenizer(br.readLine().trim());
        int[] A = new int[N];
        for (int i = 0; i < N; i++) {
            A[i] = Integer.parseInt(st.nextToken());
        }

        int sum = 0;
        int count = 0;
        int start = 0;
        int end = 0;
        // 1. sum >= M, sum 에서 A[start] 를 빼고, start 를 1 증가
        // 2. 그렇지 않으면, sum 에 end 를 더하고, end 를 1 증가
        // 3. sum == M 이면, count 를 1 증가시킵니다.
        while (true) {
            if (sum >= M) {
                sum -= A[start++];
            }
            else if (end == N) {
                break;
            } else {
                sum += A[end++];
            }
            if (sum == M) {
                count++;
            }
        }

        System.out.println(count);
    }
}
````

### 10. 연습문제: 부분합 https://www.acmicpc.net/problem/1806

#### 시간 초과 ($O(N^2)$)

````java
import java.io.BufferedReader;
import java.io.ByteArrayInputStream;
import java.io.IOException;
import java.io.InputStreamReader;
import java.util.StringTokenizer;

public class Main {
    public static void main(String[] args) throws IOException {
        String inputData = """
        10 15
        5 1 3 5 10 7 4 9 2 8
        """;
        System.setIn(new ByteArrayInputStream(inputData.getBytes()));

        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));

        StringTokenizer st = new StringTokenizer(br.readLine());

        int N = Integer.parseInt(st.nextToken());
        int S = Integer.parseInt(st.nextToken());

        st = new StringTokenizer(br.readLine());
        int[] A = new int[N];
        for (int i = 0; i < N; i++) {
            A[i] = Integer.parseInt(st.nextToken());
        }

        int min = Integer.MAX_VALUE;
        int sum = 0;

        for (int start = 0; start < N; start++) {
            for (int end = start; end < N; end++) {
                sum += A[end];
                if (sum > S) {
                    min = Math.min(min, end - start + 1);
                }
            }
            sum = 0;
        }
        System.out.println(min);
    }
}
````

#### 성공 (투포인터)

````java
import java.io.BufferedReader;
import java.io.ByteArrayInputStream;
import java.io.IOException;
import java.io.InputStreamReader;
import java.util.StringTokenizer;

public class Main {
    public static void main(String[] args) throws IOException {
        String inputData = """
        10 15
        5 1 3 5 10 7 4 9 2 8
        """;
        System.setIn(new ByteArrayInputStream(inputData.getBytes()));

        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));

        StringTokenizer st = new StringTokenizer(br.readLine());

        int N = Integer.parseInt(st.nextToken());
        int S = Integer.parseInt(st.nextToken());

        st = new StringTokenizer(br.readLine());
        int[] A = new int[N];
        for (int i = 0; i < N; i++) {
            A[i] = Integer.parseInt(st.nextToken());
        }

        int min = Integer.MAX_VALUE;
        int start = 0, end = 0, sum = 0;

        // 1. sum >=M 이면 sum 에서 A[start]를 빼고 나서 start 를 증가
        // 2. 그렇지 않다면 end 를 증가시키고 sum 에 A[end]를 더한다
        //3. sum == M 이면 count
        while (start < N && end < N) {
            if (sum >= S) {
                sum -= A[start];
                min = Math.min(end - start, min);
                start++;
            } else {
                sum += A[end];
                end++;
            }
        }

        System.out.println(min);
    }
}
````

### 11. 수 고르기 https://www.acmicpc.net/problem/2230

<div class="alert alert-block" style="border: 2px solid #E65100;background-color:#FFF3E0;padding:10px">
<font size="4em" style="color:#BF360C;">투 포인터 문제 예</font>
</div>

````java
import java.io.*;
import java.util.Arrays;
import java.util.StringTokenizer;

public class Main {

    public static void main(String[] args) throws Exception {
        String inputData = """
        3 3
        1
        5
        3
        """;
        System.setIn(new ByteArrayInputStream(inputData.getBytes()));
        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));

        StringTokenizer st = new StringTokenizer(br.readLine());

        int N = Integer.parseInt(st.nextToken());
        int M = Integer.parseInt(st.nextToken());
        int[] arr = new int[N];
        for (int i = 0; i < N; i++) {
            arr[i] = Integer.parseInt(br.readLine());
        }
        Arrays.sort(arr);

        int start = 0, end = 0, ans = Integer.MAX_VALUE;
        while (start <= end && end < N) {
            if(arr[end] - arr[start] >= M) {
                ans = Math.min(ans, arr[end] - arr[start]);
                start++;
            } else {
                end++;
            }
        }

        System.out.println(ans);
    }
}
````
